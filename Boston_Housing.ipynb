```python
import pandas as pd
data=pd.read_csv(r"/content/housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data_x=data.drop(['TAX'],axis=1)
```

```python
data_y=data['TAX']
```

```python
data_x
```

```python
data_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_x,data_y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_predicted=LR.predict(x_test)
```

```python
y_predicted
```

```python
from sklearn.metrics import mean_squared_error
```

```python
LR.score(x_test,y_test)
```

```python
mse_lr_model=mean_squared_error(y_test,y_predicted)
```

```python
mse_lr_model
```

```python
import math
rmse=math.sqrt(mse_lr_model)
```

```python
rmse
```

```python
import numpy as np
```

```python
y = data['TAX'].values
mean_y = np.mean(y)
squared_diff = (y - mean_y)**2
variance_y = np.mean(squared_diff)
print("Mean:", mean_y)
print("Variance:", variance_y)
```

```python
data.describe()
```

```python
mean_target_variable = 408.237154
normalized_mse = mse_lr_model / mean_target_variable
normalized_variance = variance_y / mean_target_variable

print("Normalized MSE:", normalized_mse)
print("Normalized Variance:", normalized_variance)
```
